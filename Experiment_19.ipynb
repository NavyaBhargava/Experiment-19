Real World and Interactive Visualizations

```python
# TreeMap
import pandas as pd
import plotly.express as px
df = pd.DataFrame({
    'Department':['HR','IT','Sales','Marketing'],
    'Budget':[20000,50000,40000,30000]
})
fig = px.treemap(df, path=['Department'], values='Budget', title ="Navya")
fig.show()
```

```python
# Dendrogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
data = np.array(
    [
        [5,3],
        [10,15],
        [15,12],
        [24,10],
        [30,30]
    ]
)
linked = linkage(data,method='ward')
plt.figure(figsize=(6,4))
dendrogram(linked)
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.title("Navya")
plt.show()
```

```python
# Venn Diagram
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
A ={1,2,3,4}
B = {3,4,5,6}
venn2([A,B],set_labels = ('A','B'))
plt.title("Navya")
plt.show()
```

```python
# Sankey Diagram
import plotly.graph_objects as go
fig = go.Figure(data=[go.Sankey(
    node=dict(
        label=['A', 'B', 'C', 'D']  # Example labels for nodes
    ),
    link=dict(
        source=[0, 1, 2],  # Indices corresponding to the 'label' array
        target=[1, 2, 3],
        value=[100, 80, 60])
)])
fig.show()
```

```python
# 3D Scatter Plot
import pandas as pd
import plotly.express as px
df = pd.DataFrame({
    'Study_Hours': [2,4,6,8,5],
    'Marks': [50,65,75,90,70],
    'Attendance':[60,75,80,90,70]
})
fig = px.scatter_3d(df, x='Study_Hours', y='Marks', z='Attendance', title='Navya')
fig.show()
```

```python
# Radar Chart
import plotly.graph_objects as go
skills = ['Python', 'ML', 'DBMS', 'DSA', 'Communication']
values = [4,3,5,4,3]
fig = go.Figure()
fig.add_trace(go.Scatterpolar(
    r=values,
    theta=skills,
    fill='toself',
    name = 'skills'
))
fig.show()
```
